```python
import torch
import torch.nn as nn
import torch.optim as optim
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from torch.utils.data import DataLoader
import tqdm as notebook_tqdm
import matplotlib.pyplot as plt
```

# Dataset

```python
diabetes_df = pd.read_csv("../dataset/diabetes.csv")
```

```python
X = diabetes_df.drop(["Outcome"], axis=1)
y = diabetes_df["Outcome"]
```

```python
sc_X = StandardScaler()
X = pd.DataFrame(sc_X.fit_transform(X), columns=X.columns)
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=42)
```

```python
def create_dataloader(x_set, y_set, batch_size):
  if torch.cuda.is_available():
    X = torch.tensor(x_set.values, dtype=torch.float32).cuda()
    y = torch.tensor(y_set.to_numpy(), dtype=torch.float32).reshape(-1, 1).cuda()
  else:
    X = torch.tensor(x_set.values, dtype=torch.float32)
    y = torch.tensor(y_set.to_numpy(), dtype=torch.float32).reshape(-1, 1)

  return DataLoader(list(zip(X, y)), shuffle=True, batch_size=batch_size)

train_loader = create_dataloader(X_train, y_train, 64)
test_loader = create_dataloader(X_test, y_test, len(X_test))
```

## **Multulayer Perceptron**

```python
class MultiLayerNet1(torch.nn.Module):
    def __init__(self, input_size, frst_lyr, scnd_lyr, third_lyr, output_size):
        super(MultiLayerNet1, self).__init__()
        self.first_hidden_layer = torch.nn.Linear(input_size, frst_lyr)
        self.second_hidden_layer = torch.nn.Linear(frst_lyr, scnd_lyr)
        self.third_hidden_layer = torch.nn.Linear(scnd_lyr, third_lyr)

        self.output_layer = torch.nn.Linear(third_lyr, output_size)

    def forward(self, x):
        frst_hidden_output = torch.relu(self.first_hidden_layer(x))
        scnd_hidden_output = torch.relu(self.second_hidden_layer(frst_hidden_output)) #, dim=1
        third_hidden_output = torch.relu(self.third_hidden_layer(scnd_hidden_output))

        y_pred = torch.sigmoid(self.output_layer(third_hidden_output))

        return y_pred
```

```python
import torch
import torch.nn as nn

class MultiLayerNet2(nn.Module):
    def __init__(self, input_size, frst_lyr, scnd_lyr, third_lyr, output_size, dropout_rate=0.7):
        super(MultiLayerNet2, self).__init__()
        
        # First hidden layer
        self.first_hidden_layer = nn.Linear(input_size, frst_lyr)
        self.dropout1 = nn.Dropout(0.3)  # Dropout

        # Second hidden layer
        self.second_hidden_layer = nn.Linear(frst_lyr, scnd_lyr)
        self.dropout2 = nn.Dropout(0.5)

        # Third hidden layer
        self.third_hidden_layer = nn.Linear(scnd_lyr, third_lyr)
        self.dropout3 = nn.Dropout(0.8)

        # Output layer
        self.output_layer = nn.Linear(third_lyr, output_size)

    def forward(self, x):
        # First hidden layer
        frst_hidden_output = torch.relu(self.first_hidden_layer(x))
        # frst_hidden_output = self.ln1(frst_hidden_output)  # Normalize output
        frst_hidden_output = self.dropout1(frst_hidden_output)  # Apply Dropout

        # Second hidden layer
        scnd_hidden_output = torch.relu(self.second_hidden_layer(frst_hidden_output))
        scnd_hidden_output = self.dropout2(scnd_hidden_output)

        # Third hidden layer
        third_hidden_output = torch.relu(self.third_hidden_layer(scnd_hidden_output))
        third_hidden_output = self.dropout3(third_hidden_output)

        # Output layer
        y_pred = torch.sigmoid(self.output_layer(third_hidden_output))  # Sigmoid for binary classification

        return y_pred
```

```python
learningRate = 0.01
epochs = 3000

model = MultiLayerNet2(8, 128, 256, 64, 1)
if torch.cuda.is_available():
    model.cuda()

criterion = torch.nn.BCELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=learningRate)
```

```python
training_loss = []
eval_loss = []

for epoch in range(epochs):
    model.train()
    for X_batch, y_batch in train_loader:
        optimizer.zero_grad()
        outputs = model(X_batch)
        loss = criterion(outputs, y_batch)
        loss.backward()
        optimizer.step()
    training_loss.append(loss.item())

    if epoch % 100 == 0:
      print('epoch {}, train loss {}'.format(epoch, loss.item()))

    # Validation
    model.eval()
    with torch.no_grad():  # Disable gradient computation during evaluation
        for x__val_batch, y__val_batch in test_loader:
            val_outputs = model(x__val_batch)
            val_loss = criterion(val_outputs, y__val_batch)
    
        if epoch % 100 == 0:
          acc = (val_outputs.round() == y__val_batch).float().mean()
          print("Model accuracy: %.2f%%" % (float(acc)*100))
          print('epoch {}, val_loss {}\n'.format(epoch, val_loss.item()))
    
        eval_loss.append(val_loss.item())
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(training_loss, linestyle='-', color='b', label='Training Loss')
plt.plot(eval_loss, linestyle=':', color='r', label='Validation Loss')

plt.xlabel('Epochs')
plt.ylabel('Cross Entropy Loss')
plt.title('Training Progress')

plt.grid(True)

# Show legend
plt.legend()
```
